```python
#  解析に必要なツールを import する.
#    pandas: データ解析を支援するライブラリ
import pandas as pd

#    numpy, matplotlib: グラフ作成を支援するライブラリ
import numpy as np
import matplotlib.pyplot as plt

#  散布図作成に必要なseaborn のimport.
import seaborn as sns
```

```python
#     データの読み取り:  AveScore.xlsx
df = pd.read_excel('AveScore.xlsx')
```

```python
plot= sns.pairplot(df)
plt.savefig('scat_ex2.png')
```
