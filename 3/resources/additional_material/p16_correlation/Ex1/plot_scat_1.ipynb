```python
#  解析に必要なツールを import する.
#    pandas: データ解析を支援するライブラリ
import pandas as pd

#    numpy, matplotlib: グラフ作成を支援するライブラリ
import numpy as np
import matplotlib.pyplot as plt

#  散布図作成に必要なseaborn のimport.
import seaborn as sns
```

```python
#     データの読み取り:  NumWin.xlsx
data= pd.read_excel('NumWin.xlsx' )
```

```python
#  散布図の作成
plot =  sns.jointplot(x='Average_score', y='Number_of_wins',data= data, kind='scatter')
plot.ax_marg_x.set_xlim([3,6])
plot.ax_marg_x.set_xticks([ 3, 4, 5, 6])
plot.ax_marg_y.set_ylim([50,100])
plt.savefig('scat_ex1.png')
```
