```python
#  グラフ作成に必要なツールを import する.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#  データの読み取り: Tokyo_month.xlsx
data = pd.read_excel('Tokyo_month.xlsx', index_col=0)
#  データの整形
data.index = data.index.strftime('%Y/%m')
```

```python
# 折れ線グラフの作成
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.set_xticks([0, 240, 480, 720, 960, 1200, 1440])

data.plot(ax=ax)

#  凡例を削除
leg = plt.legend()
ax.get_legend().remove()

plt.savefig('Tokyo_month_all.png')
```
