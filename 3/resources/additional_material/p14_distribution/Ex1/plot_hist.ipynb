```python
#  ライブラリの読み込み
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#     データ (weight.xlsx) を読み込む.
data= pd.read_excel('weight.xlsx',  header=None)
data
```

```python
#  ヒストグラムの作成 (1)
plt.hist(data[1], bins= 12, range=(60, 120) )
plt.xlabel("Weight (Kg)")
plt.ylabel("Count")
plt.savefig('hist.png')
```

```python
#  ヒストグラムの作成 (2):  ビン幅が太すぎる
plt.hist(data[1], bins= 2, range=(60, 120) )
plt.xlabel("Weight (Kg)")
plt.ylabel("Count")
plt.savefig('hist_large.png')
```

```python
#  ヒストグラムの作成 (3):  ビン幅が細すぎる
plt.hist(data[1], bins= 60, range=(60, 120) )
plt.xlabel("Weight (Kg)")
plt.ylabel("Count")
plt.savefig('hist_small.png')
```
