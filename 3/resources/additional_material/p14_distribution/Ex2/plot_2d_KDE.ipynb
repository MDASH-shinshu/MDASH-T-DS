```python
#  ライブラリの読み込み
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#     データ (Height_Weight.xlsx) を読み込む.
data= pd.read_excel('Height_Weight.xlsx',  header=None)
data.columns = ["name","Weight","Height"]
data
```

```python
# 2次元 KDE の作成
plot =  sns.jointplot(x="Weight", y="Height",data= data, kind='kde')
plt.savefig('2d_kde.png')
```
