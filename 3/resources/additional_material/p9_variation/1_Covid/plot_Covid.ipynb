```python
#  解析に必要なツールを import する.
#    pandas: データ解析を支援するライブラリ
import pandas as pd

#    numpy, matplotlib: グラフ作成を支援するライブラリ
import numpy as np
import matplotlib.pyplot as plt
```

```python
#     データ (covid_cases.xlsx) を読み込む.
data = pd.read_excel("covid_cases.xlsx", index_col= 0)
data.index = data.index.strftime('%Y/%m/%d')
```

```python
#　　Figureを設定
fig = plt.figure()

#　　Axesを追加
ax = fig.add_subplot(1, 1, 1)
ax.plot(data)


#　　軸ラベルの設定
plt.xlabel("day")
plt.ylabel("cases")

#　　軸の目盛りの設定
plt.xticks([15, 103, 195, 286, 376])

#　　グラフを保存
fig.savefig('covid.png')
```
